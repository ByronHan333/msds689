```python
class Node:
  def __init__(self, value):
    self.value = value
    self.edges = [] # outgoing edges
  def add(self, target):
    self.edges.append(target)
  def __repr__(self): return str(self.value)
```

```python
sf = Node("SF")
la = Node("LA")
sac = Node("Sacramento")
oak = Node("Oakland")
baker = Node("Bakersfield")
sj = Node("San Jose")
sf.add(sj)
sj.add(baker)
sf.add(oak)
oak.add(sac)
sac.add(baker)
baker.add(la)

objviz(sf)
```

# Reachable nodes

```python
def reachable(p:Node) -> set:
    reaches = set();
    reachable_(p, reaches, set())
    return reaches

def reachable_(p:Node, reaches:set, seen:set):
    if p in seen: return
    seen.add(p)
    for q in p.edges:
        reaches.add(q)
        reachable_(q, reaches, seen)
```

```python
reachable(sf)
```

```python
reachable(sj)
```

```python
reachable(baker)
```
